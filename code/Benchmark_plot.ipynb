```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Triplet_hardmining_10k = np.load('Triplet_hardmining_10k.npy')
Triplet_withclass_66k = np.load('Triplet_withclass_66k.npy')
Siamese_6k =np.load('Siamese_6k.npy')
Siamese_classificaion_only = np.load('Siamese_classificaion_only.npy')
Alexnet_canny_extract = np.load('Alexnet_canny_extract.npy')
Sketchnet_canny_extract = np.load('Sketchnet_canny_extract.npy')
SHOG = np.load('shog.npy')
GFHOG = np.load('gfhog.npy')
Random = np.load('Random.npy')
```
