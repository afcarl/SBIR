```python
NUM_PER_CLASS = 1000 #number of images per class, change this

query_list = np.copy(cat_list)
query_list[12] = "car"
query_list[2] = "apple,fruit"

for i,query in enumerate(query_list[:30]):
    if i==0 or i==1:
        continue
    w = Walker(Photo.search, tags=query,tag_mode="all")
    usr = ""
    count_class = 0
    num_per_class = NUM_PER_CLASS - np.shape(os.listdir("flickr_add/"+cat_list[i]))[0]
    print "\nRetrieving "+str(num_per_class)+" for query: " + query
    for photo in w:
        if count_class >= num_per_class:
            break
        try:
            if photo['owner'].username == usr:
                continue
        except:
            continue
        usr = photo['owner'].username
        save_path = "flickr_add/"+cat_list[i]+"/flickr_"+photo['id']+'.jpg'
        if os.path.isfile(save_path):
            continue
        try:
            print '\r',
            print "\t "+str(count_class)+" saving "+save_path+"... ",
            photo.save(save_path,size_label="Medium")
            print "done",
        except:
            print "failed",
            continue
        count_class = count_class+1
```
